```python
import subprocess
import sys
import os

# Path to the folder containing your scripts
BASE_DIR = ""#"Project/Project/IntermediaryPythonFiles"

# Scripts to run in order
scripts = [
    "Fetch_Data.py",
    "Train.py",
    "TrainSKLearnLog.py",
    "TrainXGBoost.py",
    "X_SelectBestModelAndRegister.py",
    "Z_Deploy.py"
]

for script in scripts:
    full_path = os.path.join(BASE_DIR, script)
    print(f"\n=== Running {script} ===\n")
    result = subprocess.run([sys.executable, full_path])
    
    if result.returncode != 0:
        print(f"{script} failed with exit code {result.returncode}")
        break
    
    print(f"âœ” Finished {script}")
```
