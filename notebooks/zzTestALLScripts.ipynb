```python
import subprocess
import sys
import os
import datetime


# Path to the folder containing your scripts
#BASE_DIR = "../../Module/pythonFiles"#"Project/Project/IntermediaryPythonFiles"
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), "..", "Module", "pythonFiles"))


# Constants used:
current_date = datetime.datetime.now().strftime("%Y_%B_%d")
experiment_name = current_date


# Scripts to run in order
scripts = [
    #"Fetch_Data.py",
    
    "1.0_Train.py",
    "1.1_TrainXGBoost.py",
    "1.2_TrainSKLearnLog.py",
    "2.0_SelectBestModelAndRegister.py",
    "3.0_Deploy.py"
]

for script in scripts:
    full_path = os.path.join(BASE_DIR, script)
    print(f"\n=== Running {script} ===\n")

    result = subprocess.run(
        [sys.executable, full_path, "--run_name", experiment_name]
    )

    if result.returncode != 0:
        print(f"{script} failed with exit code {result.returncode}")
        break

    print(f"✔ Finished {script}")
```
